- ladder (2x)
- next coin (2x)

therefore there are 8 possible paths

```python
from typing import List
from pandas import DataFrame
from numpy import inf

def get_path(s, dist):
    # returns the distance for a given path and starting point
    total_distance = 0
    for d in dist:
        if d is None:
            continue
        print(s)
        total_distance += abs(d-s)
        s = d
    return total_distance

def get_min_path(sx, x_min, x_max, lx1, lx2, nx_min, nx_max):
    #check all 8 paths
    
    min_d = inf
    temp_sx = 0
    
    #path1
    d = get_path(sx, [x_min, x_max, lx1, nx_min])
    if d < min_d:
        min_d = d
        temp_sx = nx_min
        
    #path2
    d = get_path(sx, [x_max, x_min, lx1, nx_min])
    if d < min_d:
        min_d = d
        temp_sx = nx_min
        
    #path3
    d = get_path(sx, [x_min, x_max, lx2, nx_min])
    if d < min_d:
        min_d = d
        temp_sx = nx_min
        
    #path4
    d = get_path(sx, [x_max, x_min, lx2, nx_min])
    if d < min_d:
        min_d = d
        temp_sx = nx_min
        
    #path5
    d = get_path(sx, [x_min, x_max, lx1, nx_max])
    if d < min_d:
        min_d = d
        temp_sx = nx_max
        
    #path6
    d = get_path(sx, [x_max, x_min, lx1, nx_max])
    if d < min_d:
        min_d = d
        temp_sx = nx_max
        
    #path7
    d = get_path(sx, [x_min, x_max, lx2, nx_max])
    if d < min_d:
        min_d = d
        temp_sx = nx_max
        
    #path8
    d = get_path(sx, [x_max, x_min, lx2, nx_max])
    if d < min_d:
        min_d = d
        temp_sx = nx_max
    
    return d, temp_sx

def minSteps(sx: int, lx1: int, lx2: int, X: List[int], Y: List[int]) -> int:
    t = 0
    y = 0
    n_collected = 0
    
    df = DataFrame({'x':X, 'y':Y}).sort_values(['y', 'x'], ignore_index=True)
    display(df)
    
    while True:
        
        #y-th floor
        mask1 = df['y'] == y
        
        #no coin in y-th floor
        if mask1.sum() == 0:
            x_min, x_max = None, None
        else:
            x_min, x_max = df[mask1]['x'].min(), df[mask1]['x'].max()
        
        #next floor
        mask2 = df['y'] > y
        
        #if last floor
        if mask2.sum() == 0: 
            nx_min, nx_max = None, None
            lx1, lx2 = None, None
            dy = 0
        else:
            n_y = df[mask2].iloc[0]['y']
            nx_min, nx_max = df[df['y'] == n_y]['x'].min(), df[df['y'] == n_y]['x'].max()
            dy = n_y-y
        
        d, sx = get_min_path(sx, x_min, x_max, lx1, lx2, nx_min, nx_max)
        
        t += d
        y += dy
        
        if mask2.sum() == 0:
            break
        
    return t

def test():
    assert (a := minSteps(10, 0, 100, [47, 42], [0, 0])) == 37, a
    assert (a := minSteps(10, 8, 11, [10, 12], [1, 1])) == 5, a
    assert (a := minSteps(10, 8, 42, [10, 12], [1, 1])) == 7, a
    assert (a := minSteps(10, 8, 42, [10, 100, 12], [1, 0, 1])) == 181, a
    assert (a := minSteps(
                500000000, 1, 999999999, 
                [0, 1000000000, 0, 1000000000, 0, 1000000000, 0, 1000000000, 0, 1000000000, 0, 1000000000, 0, 1000000000],
                [1, 1, 2, 2, 3, 3, 4, 4, 7, 7, 999999947, 999999947, 900000047, 900000047]
            )
    ) == 8499999959, a
    
    print('all good')
test()
```
